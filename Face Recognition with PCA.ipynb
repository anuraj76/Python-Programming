```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
X = []
y = []
```

```python
#loading Images into numpy arrays

for dir_path,dir_names, file_names in walk('D:\ML and AI PGPG\Mastering Machine Learning\9781788621113_Code (1)\Mastering ML with Sckit learn\9781783988365_Code\8365OS__CodeBundle\8365OS_07_Codes'):
    for fn in file_names:
        if fn[-3:] == 'pgm':
            image_filename = path.join(dir_path, fn)
            X.append(scale(cv2.imread(image_filename,0).reshape(10304).astype('float32')))
            y.append(dir_path)
```

```python
X = np.array(X)
```

```python
X
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3)
```

```python
pca = PCA(n_components=110)
```

```python
Xtrain_reduced = pca.fit_transform(Xtrain)
```

```python
print(Xtrain.shape)
print(Xtest.shape)
print(Xtrain_reduced.shape)
```

```python
Xtest_reduced = pca.fit_transform(Xtest)
```

```python
classifier = LogisticRegression()
```

```python
accuracies = cross_val_score(classifier,Xtrain_reduced,Ytrain)
```

```python
print('Cross Vlaidation accuracy:',np.mean(accuracies))
```

```python
classifier.fit(Xtrain_reduced,Ytrain)
```

```python
prediction = classifier.predict(Xtest_reduced)
```

```python
print(classification_report(Ytest,prediction))
```
