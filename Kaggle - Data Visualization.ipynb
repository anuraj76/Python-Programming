```python
df_flightsdelays = pd.DataFrame(pd.read_csv('datasets_116573_334386_flight_delays.csv',index_col='Month'))
```

```python
df_flightsdelays
```

```python
# Bar chart for  average arrival delay for Spirit Airlines (airline code: NK) flights, by month.
```

```python
plt.figure(figsize=(16,4))
plt.title('Avergare Arrival Flight Delay for Spirit Airways, by Month')
sns.barplot(x=df_flightsdelays.index,y=df_flightsdelays['NK'])
plt.ylabel('Delay in Mins')

```

```python
#Plottype - HeatMap
```

```python
plt.figure(figsize=(16,7))
plt.title('Avg Arrival delay for each Airline, by Month')
sns.heatmap(data=df_flightsdelays, annot=True)
plt.xlabel('Airline')
```

```python
sns.scatterplot(x=df_insurance['bmi'],y=df_insurance['charges'],hue=df_insurance['smoker'])
```
