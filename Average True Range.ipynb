```python
import numpy as np
```

```python
import sys
```

```python
h,l,c = np.loadtxt('data.csv',delimiter=',',usecols=(4,5,6), unpack=True)
```

```python
N = 5
h = h[-N:]
l = l[-N:]
```

```python
print("Len h : ",len(h),"Len l : ", len(l) )
```

```python
print("Close" , c)
```

```python
previousclose = c[-N -1: -1]
```

```python
print("Previousclose:",previousclose)
```

```python
truerange = np.maximum(h-1,h-previousclose,previousclose-l)
```

```python
print("Truerange :",truerange)
```

```python
atr = np.zeros(N)
```

```python
atr[0] = np.mean(truerange)
```

```python
for i in range(1,N):
    atr[i] = (N - 1) * atr[i - 1] + truerange[i]
    atr[i] /= N
print("ATR", atr)    
```
