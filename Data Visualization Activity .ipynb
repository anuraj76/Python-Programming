```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('mtcars.csv')
```

```python
cars
```

```python
plt.figure(figsize=(10,4))
WT = cars['wt']
MPG = cars['mpg']
plt.scatter(WT,MPG)
```

```python
WT = cars['wt']
DRAT = cars['drat']
Cars_data = pd.DataFrame({'Car Weight':WT,'Car Drat':DRAT})
Cars_data
```

```python
for i in Cars_data.columns:
    plt.plot(Cars_data[i])
    plt.title('Car Details for:'+ i)
    plt.show()
```

```python
# Activity 2 - use HSB data to populate the line series for read,write, math, science and socst variables through for loop
```

```python
HSB = pd.read_excel('hsb.xls')
```

```python
HSB
```

```python
Math = HSB['math'][1:10]
Science = HSB['science'][1:10]
READ = HSB['read'][1:10]
WRITE = HSB['write'][1:10]
SOCST = HSB['socst'][1:10]
```

```python
HSB_data = pd.DataFrame({'Read':READ,'Write':WRITE,'Math':Math,'Science':Science,'SOcst':SOCST})
```

```python
HSB_data
```

```python
for i in HSB_data.columns:
    plt.plot(HSB_data[i])
    plt.title('HSB_data for:'+ i)
    plt.show()
```
