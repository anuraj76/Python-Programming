```
import cv2
```

```
i = cv2.imread('/content/drive/My Drive/Colab Notebooks/DatasetALandP/alien_vs_predator_thumbnails/data/train/alien/80.jpg')
```

```
from google.colab.patches import cv2_imshow

cv2_imshow(i)
```

```
# Averaging

# Def a nxn sliding window on top of our image
import numpy as np


blurred = np.hstack([
                     cv2.blur(i,(3,3)),
                     cv2.blur(i,(5,5)), cv2.blur(i,(7,7))
])
```

```
cv2_imshow(blurred)
```

```
#Guassian
blurred = np.hstack([
                     cv2.GaussianBlur(i,(3,3),0),
                     cv2.GaussianBlur(i,(5,5),0), cv2.GaussianBlur(i,(7,7),0)
])
```

```
cv2_imshow(blurred)
```

```
#Median
blurred = np.hstack([
                     cv2.medianBlur(i,3),
                     cv2.medianBlur(i,5), cv2.medianBlur(i,7)
])

cv2_imshow(blurred)
```

```
#bilateral In order to reduce noise while still maintaining edges, we can use bilateral blurring

blurred = np.hstack([
                     cv2.bilateralFilter(i,5,21,21),
                     cv2.bilateralFilter(i,7,31,31), cv2.bilateralFilter(i,9,41,41)
])

cv2_imshow(blurred)
```

```
#Thresholding - is the binarization of an image
#to convert a grayscale image to a binary image,where the pixels are either 0 or 255


image = cv2.imread('/content/drive/My Drive/Colab Notebooks/DatasetALandP/alien_vs_predator_thumbnails/data/train/alien/90.jpg')
```

```
cv2_imshow(image)
```

```
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```
cv2_imshow(image)
```

```
blurred  = cv2.GaussianBlur(image,(5,5),0)
```

```
cv2_imshow(blurred)
```

```
(T,thresh) = cv2.threshold(blurred,155,255,cv2.THRESH_BINARY)
cv2_imshow(thresh)
```

```
(T,threshinv) = cv2.threshold(blurred,155,255,cv2.THRESH_BINARY_INV)
cv2_imshow(threshinv)
```

```
cv2_imshow(cv2.bitwise_and(image,image,mask=threshinv))
```

```
#Adaptive thresholding

thresh = cv2.adaptiveThreshold(blurred,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,11,4)
cv2_imshow(thresh)
```

```

thresh = cv2.adaptiveThreshold(blurred,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,4)
cv2_imshow(thresh)
```
